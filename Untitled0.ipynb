```python
# SIMPLE ALGERIA INFLATION ANALYSIS
# BY ABDENNOUR HEMICI

print("=" * 50)
print("ALGERIA INFLATION ANALYSIS - NOVEMBER 2025")
print("=" * 50)
print()

# Data from ONS report
categories = ["Food", "Clothing", "Housing", "Transport", "Jewelry"]
changes = [-0.67, 0.43, 0.00, 0.00, 10.74]

# Show data
print("PRICE CHANGES (%):")
for i in range(len(categories)):
    print(f"{categories[i]}: {changes[i]}%")
print()

# Find highest/lowest
highest = max(changes)
lowest = min(changes)
highest_category = categories[changes.index(highest)]
lowest_category = categories[changes.index(lowest)]

print("ANALYSIS RESULTS:")
print(f"Highest increase: {highest_category} (+{highest}%)")
print(f"Highest decrease: {lowest_category} ({lowest}%)")
print(f"Average change: {sum(changes)/len(changes):.2f}%")
print()

# Simple chart
print("VISUALIZATION (text version):")
for i in range(len(categories)):
    bar = "█" * int(abs(changes[i]) * 2)  # Simple bar
    if changes[i] >= 0:
        print(f"{categories[i]:15} [+]{bar}")
    else:
        print(f"{categories[i]:15} [-]{bar}")
```
